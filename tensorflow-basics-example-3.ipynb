* the scaler should be of dtype "int32"
* the vector should be of the shape (3).and type of dtype "int32"
* Assign the scaler to the variable `"scaler"`

# Question 3:
### perform the following operations
* Create a random tensor from a uniform distribution of shape (5,3,3,2) with values between 0 and 1 with the seed value as 32 and assign the tensor to the variable "random_1"
* Create a random tensor from a uniform distribution of shape (5,3,3,2) with values between 0 and 1 and assign the tensor to the variable "random_1"

```python
def random_tensor():
    tf.random.set_seed(32)
    random_1=tf.random.uniform(shape=(5,3,3,2),minval=0,maxval=1)
    random_2=tf.random.uniform(shape=(5,3,3,2),minval=0,maxval=1)
    return [random_1,random_2]
```

```python
random_tensor=random_tensor()
random_tensor
```

# Question 4 :
### Perform matrix multiplication
#### Given two matrices matrix1,matrix2 with identical shape (5,3), find a way to perform matrix multiplication with resultant shape of (5,5)
* Assign the resultant vector to the variable "resultant_vector"

```python
def matrix_multiplication(matrix1,matrix2):
    resultant_vector= tf.matmul(matrix1,matrix2)
    return resultant_vector
```

```python
matrix1=tf.constant([[1,2,3],[3,4,5],[5,6,7],[7,8,9],[9,10,11]])
matrix2=tf.constant([[11,12,13,14,15],[15,16,17,18,19],[19,20,21,22,23]])
matrix_multiplication=matrix_multiplication(matrix1,matrix2)
matrix_multiplication
```

# Question 5:
### Aggregation , find the followings for the given `"tensor_1"`
* minimum and assign the value to the variable `"min_val"`
* maximum and assign the value to the variable `"max_val"`
* mean and assign the value to the variable `"mean_val"`
* sum and assign the value to the variable `"sum_val"`

```python
def perform_aggregate(tensor_1):
    min_val=tf.reduce_min(tensor_1)
    max_val=tf.reduce_max(tensor_1)
    mean_val=tf.reduce_mean(tensor_1)
    sum_val=tf.reduce_sum(tensor_1)
    return [min_val,max_val,mean_val,sum_val]
```

```python
tensor_1=tf.constant([ [[1,2,3],[2,3,4],[4,5,6]] ,[[6,7,8],[8,9,10],[10,11,12]], [[12,13,14],[14,15,16],[16,17,18]], [[18,19,20],[20,21,22],[22,23,24]] ])
perform_aggregate=perform_aggregate(tensor_1)
perform_aggregate
```

# Question 6 :
### Find the minimum and maximum element index of each vector for the given tensor `"tensor_1'`
* Assign the maximum element indexes to the variable `"max_pos"`
* Assign the minimum element indexes to the variable `"max_pos"`
* The indexes should be of the type `"numpy array"`

```python
def min_max_position(tensor_1):
    max_pos = tf.argmax(tensor_1,axis=-1)
    min_pos = tf.argmin(tensor_1,axis=-1)
    return [max_pos,min_pos]
```

```python
tensor_1=tf.constant([ [[1,2,3],[2,3,4],[4,5,6]] ,[[6,7,8],[8,9,10],[10,11,12]], [[12,13,14],[14,15,16],[16,17,18]], [[18,19,20],[20,21,22],[22,23,24]] ])
min_max_position=min_max_position(tensor_1)
min_max_position
```

# Question 7:
### one hot encode the given tensor `"onehot_tensor"` and assign  then encoded tensor to the variable `"encoded_tensor"`
* the input tensor might contain values between 0-9
* the returned tensor should be of dtype "int32"

```python
def one_hot_encode_tensor(onehot_tensor):
    depth=10
    encoded_tensor = tf.one_hot(onehot_tensor,depth,dtype=tf.int32)
    return encoded_tensor
```

```python
onehot_tensor = [0, 2, 4, 1, 3, 5, 0, 2, 9]
one_hot_encode_tensor = one_hot_encode_tensor(onehot_tensor)
one_hot_encode_tensor
```
