# What is the difference between 50th percentile and 75th percentile of average temperature.

```python
avg_50_75_diff =climate['Average temperature (°F)'].quantile(q=0.75)-climate['Average temperature (°F)'].quantile(q=0.5)
print('avg_50_75_diff :', round(avg_50_75_diff,2))
```

# What is the pearson's correlation between average dew point and average temperature.

# Which month has the highest median for maximum_gust_speed out of all the available records.Also find the respective value( hint: group by month)

```python
# Method 1 :- 
from datetime import datetime
climate['Date']=[datetime.strptime(d,"%d/%m/%Y") for d in climate['Day']]  # necessary to convert in datetime. pd.to_datetime directly won't provide correct answers.
climate['Date1']=pd.to_datetime(climate['Date'])# necessary to access month
climate['month']=climate.Date1.dt.month
# Method 2:-
# climate['Date']=pd.to_datetime(climate['Day'])
# climate['month']=climate.Date.dt.month

hig_max_gst_spd =climate.groupby('month')['Maximum gust speed (mph)'].median().max()
hig_max_gst_spd_mont=climate.groupby('month')['Maximum gust speed (mph)'].median().idxmax()
print("hig_max_gst_spd : " , hig_max_gst_spd)
print("hig_max_gst_spd_mont : " , hig_max_gst_spd_mont)
```

# Determine the average temperature between the months of March 2010 to May 2012 (including both the months)

```python
val=climate[(climate['Date1']>='2010-03-01')& (climate['Date1']<='2012-05-31')]['Average temperature (°F)'].mean()
print('avg_temp_mar_may :' ,round(val,2) )
```

# Find the range of average temperature on Dec 2010

```python
min_temp=climate[(climate['Date1']>='2010-12-01') & (climate['Date1']<='2010-12-31')]['Average temperature (°F)'].min()
max_temp=climate[(climate['Date1']>='2010-12-01') & (climate['Date1']<='2010-12-31')]['Average temperature (°F)'].max()
range1=max_temp-min_temp
print("rang_avg_temp : ",round(range1,2))
```

# Out of all the available records how many days are within one standard deviation of average temperature.
* "Within one standard deviation of the mean" means within the interval [x¯−σ,x¯+σ]
* x¯ - mean and σ - standard deviation

```python
x_mean=climate['Average temperature (°F)'].mean()
x_std=climate['Average temperature (°F)'].std()
one_std_low= round((x_mean - x_std),2)
one_std_hig= round((x_mean + x_std),2)
climate_one_std_rannge=climate[(climate['Average temperature (°F)']>=one_std_low)& (climate['Average temperature (°F)']<=one_std_hig)]
print("avg_withn_val :",len(climate_one_std_rannge))
```
