# Tensors - Scaler , Vector And Matrix

```python
Scaler=tf.constant([2])
Vector=tf.constant([5,6,2])
Matrix=tf.constant([[1,2,3],[2,3,4],[3,4,5]])
Tensor=tf.constant([ [[1,2,3],[2,3,4],[3,4,5]] , [[4,5,6],[5,6,7],[6,7,8]] , [[7,8,9],[8,9,10],[9,10,11]] ])

with tf.Session() as session:
    scaler_value = session.run(Scaler)
    vector_value = session.run(Vector)
    matrix_value = session.run(Matrix)
    tensor_value = session.run(Tensor)
    # Print the values
    print("Scaler (1 entry):\n", scaler_value)
    print("\nVector (3 entries):\n", vector_value)
    print("\nMatrix (3x3 entries):\n", matrix_value)
    print("\nTensor (3x3x3 entries):\n", tensor_value)
    
# print("Scaler (1 entry):\n",Scaler.numpy())
# print("\nVector (3 entries):\n",Vector.numpy())
# print("\nMatrix (3x3 entries):\n",Matrix.numpy())
# print("\nTensor (3x3x3 entries):\n",Tensor.numpy())
```

```python
Matrix_one = tf.constant([[1,2,3],[2,3,4],[3,4,5]])
Matrix_two = tf.constant([[2,2,2],[2,2,2],[2,2,2]])

first_operation=tf.add(Matrix_one,Matrix_two)
second_operation=Matrix_one+Matrix_two
third_operation=tf.matmul(Matrix_one,Matrix_two)

with tf.Session() as session:
    first_operation = session.run(first_operation)
    second_operation = session.run(second_operation)
    third_operation = session.run(third_operation)
    print("Using TensorFlow functions:\n",first_operation)
    print("\nUsing normal experssion :\n",second_operation)
    print("\nUsing TensorFlow matmul functions :\n",third_operation)


# print("Using TensorFlow functions:\n",first_operation.numpy())
# print("\nUsing normal experssion :\n",second_operation.numpy())
# print("\nUsing TensorFlow matmul functions :\n",third_operation.numpy())
```

# Constant , Variables And Placeholders

#### Constants:
* Constants are tensors whose values cannot be changed. They are typically used to hold fixed values that are known before the graph is executed.

#### Variables:
* Variables are tensors whose values can be changed. They are typically used to hold trainable parameters of a model, such as weights and biases.
* Before using variables, you must initialize them.

#### Placeholders:
* Placeholders are tensors used to feed external data into the TensorFlow graph. They are typically used for inputs to the model.
* Unlike constants and variables, placeholders do not have initial values and do not hold any data.They serve as a placeholder for data that will be provided later during the execution of the graph.

```python
#Variables
tf.disable_v2_behavior()

one=tf.constant(1)
state=tf.Variable(0, dtype=tf.int32)

new_value=tf.add(state,one)
update=tf.assign(state,new_value) # tf.assign() is depreciated in TensorFlow 2.x.

init_op=tf.global_variables_initializer() #tf.intialize_all_variables() is depreciated in TensorFlow 2.x.

# Run the session
with tf.Session() as session: #Session is deprecated in TensorFlow 2.x . So using compat.v1 for version compability for TensorFlow 1.x
    # Initialize variables
    session.run(init_op)
    # Update the state variable
    for _ in range(3):
        session.run(update)
        print(session.run(state))
```

```python
#placeholders
a=tf.placeholder(tf.float32)
b=a*2
```

```python
with tf.Session() as session:
    result=session.run(b,feed_dict={a:3.5})
    print(result)
```
