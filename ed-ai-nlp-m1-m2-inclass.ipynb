### Word Tokenization

```python
#Toeknization
#word level toeknization -- split into words
import nltk
from  nltk.tokenize import word_tokenize
tokenized_string = word_tokenize(Text[0])
print("actual_string :", Text[0])
print("tokenized_string:" ,tokenized_string)
```

### Frequency Distribution

```python
## what are tope words mentioned in the data?
from nltk.probability import FreqDist
fdist = FreqDist()
for word in tokenized_string:
  fdist[word.lower()]+=1
fdist
```

```python
# Complete  Caption tokenization
tokenized_full_string=[]
for text in Text:
  tokenized_full_string .extend(word_tokenize(text))
#print(tokenized_full_string)
fdist_full = FreqDist()
for word in tokenized_full_string:
  fdist_full[word.lower()]+=1
fdist_full
```

### Regex Tokenization

```python
#regex tokenizer
# regular experssion are used to extract /match/substitute  certeian pattern in data
# https://regex101.com/    -- to get the pattern accroding to need
sample_text = 'Sun Pharmaceutical Industries Ltd on Tuesday said it will acquire a 16.33 per cent stake in Surgimatix, Inc, a US-based firm for $3.05 million (over Rs 25 crore). Sun Pharmaceutical has signed a definitive merger agreement for the acquisition of all outstanding ordinary shares of Taro Pharmaceutical not currently owned by the company or its affiliates for $347.73m (Rs28. 92bn) in cash.'
from nltk.tokenize import regexp_tokenize
regexp_tokenize(sample_text,pattern = '\d+\.?\d*')
#explanation
# \d -- [0 to 9]
#  \.? -- decimal can be after digit or not that the reason '?'
# \d* -- after decimal it could be zero or more
```

### Sentence Tokenization

```python
## sentence tokenization -- split into sentences
from nltk.tokenize import sent_tokenize
sent_tokenize(sample_text)
```

### Blankline Tokenization

```python
## blankline tokenization
from nltk.tokenize import blankline_tokenize
blankline_tokenize(sample_text)
```

## Bigrams, Trigrams & Ngrams

```python
# unigram - Single word
# bigram - two word
# trigram - three word
from nltk.util import bigrams,trigrams,ngrams
#unigrams
sample_text_tokens = word_tokenize(sample_text)
print(sample_text_tokens)
```

```python
#bigrans
sample_text_bigrams = list(bigrams(sample_text_tokens))
print(sample_text_bigrams)
```

```python
#trigrams
sample_text_trigrams = list(trigrams(sample_text_tokens))
print(sample_text_trigrams)
```

```python
#ngrams
sample_text_ngrams= list(ngrams(sample_text_tokens,4))
print(sample_text_ngrams)
```

## Stemming
 Process of normalizing words into the root form



### PorterStemmer

```python
# Stemming : process of normalizing words into the root form

## S1: I am watching a movie and it is awesome : positive
## S2: I watched this movie and it was awesome : positive

## watched , watching - watch root form
from nltk.stem import PorterStemmer  # removes 'ies','ied','s','tion','test','ying'
pst =  PorterStemmer()
print(pst.stem("having"))
print(pst.stem("studies"))
print(pst.stem("fishing"))
print(pst.stem('give'))
print(pst.stem('giving'))
print(pst.stem('given'))
print(pst.stem('gave'))
```

### LancasterStemmer

```python
from nltk.stem import LancasterStemmer  # removes 'ies','ied','s','tion','test','ying'
lst =  LancasterStemmer()
print(lst.stem("having"))
print(lst.stem("studies"))
print(lst.stem("fishing"))
print(lst.stem('give'))
print(lst.stem('giving'))
print(lst.stem('given'))
print(lst.stem('gave'))
```

### SnowBallStemmer

```python
from nltk.stem import SnowballStemmer  # removes 'ies','ied','s','tion','test','ying'
sbt = SnowballStemmer('english')
sbt.languages  ## supporting languages
```

```python
print(sbt.stem("having"))
print(sbt.stem("studies"))
print(sbt.stem("fishing"))
print(sbt.stem('give'))
print(sbt.stem('giving'))
print(sbt.stem('given'))
print(sbt.stem('gave'))
```
