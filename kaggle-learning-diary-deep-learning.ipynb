 > **Stemming** : process of removing suffixes (and sometimes prefixes) from words to return them to their root form, known as the "stem."
* **PorterStammer** : It uses a series of rules to iteratively reduce words to their stems.

    `from nltk.stem import PorterStemmer # removes 'ies','ied','s','tion','test','ying'
    pst =  PorterStemmer()
    print(pst.stem("having"))`
* **LancasterStemmer** : Known for being more aggressive in its reductions compared

    `from nltk.stem import LancasterStemmer  # removes 'ies','ied','s','tion','test','ying'
    lst =  LancasterStemmer()
    print(lst.stem("having"))`

* **SnowBallStemmer** :

    `from nltk.stem import SnowballStemmer  # removes 'ies','ied','s','tion','test','ying'
    sbt = SnowballStemmer('english')
    print(sbt.stem("having"))`

* **TF-IDF** :
    * Term Frequency (TF): 
    TF= (Total number of terms in the document) / (Number of times a term appears in a document)
    * Inverse Document Frequency (IDF) :
    IDF=log(Number of documents containing the term / Total number of documents)
    * TF-IDF Score: The product of TF and IDF.
        * TF-IDF=TF×IDF
        
`from sklearn.feature_extraction.text import TfidfVectorizer
vectorize_tfidf=TfidfVectorizer(stop_words="english",   #stopwords considered as english
                                max_df=0.98, #remove a word which repeated 98% in the documents
                                min_df=1, # remove a word no repeeation
                                max_features=None)
